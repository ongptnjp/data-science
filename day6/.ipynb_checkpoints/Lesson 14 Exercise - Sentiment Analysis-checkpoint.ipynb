# 2. Specify labels




```python
print (data.text[data.label == 5].sample(1).tolist()[0])
```

```python
def make_labels(star):
    star = int(star)
    if star > 3:
        return 'positive'
    elif star == 3:
        return 'neutral'
    else:
        return 'negative'

print (data.label.apply(make_labels).value_counts())
```

# 3. Clean text และ Tokenize text


```python
import pythainlp 
tokenized_text = [pythainlp.tokenize.word_tokenize(x, engine='attacut') for x in data.iloc[0:10, 0]]
```

Tokenized data:

https://drive.google.com/file/d/1TNuPJP61ss_kb80CWd57GfRWxEh57ssY/view?usp=sharing

```python
t_data = pd.read_csv('./wongnai_tokenized.csv')
tokenized_text = t_data.tokens.apply(lambda x: x.split('|'))
```

```python
t_data.head()
```

```python
stopwords = 
```

```python
import re
def clean_text(token_list):
    patt = re.compile('[ก-๙0-9a-zA-Z]+')
    cleaned_token_lists = []
    for row in token_list: # แต่ละข้อความ
        cleaned_tokens = []
        for w in row: # แต่ละคำในข้อความ 
            if patt.match(w):
                cleaned_tokens.append(w)
        cleaned_token_lists.append(cleaned_tokens)
    return cleaned_token_lists
    #cleaned_text = [[x for x in row if patt.match(x)] for row in token_list]
```

```python
cleaned_text = clean_text(tokenized_text)
```

```python
'|'.join(cleaned_text[12])
```

```python
len(cleaned_text)
```

# 4. กำหนดค่าสัญญาณ (features) 





```python
import collections
c = collections.Counter()
for row in tokenized_text:
    for word in row:
        c[word] += 1
```

```python
stopwords = set()
for word, _ in c.most_common(100):
    stopwords.add(word)
```

```python
def featurize(tokenized_text):
    feature_dict = {}
    for word in tokenized_text:
        #if word not in stopwords:
        feature_dict[word] = 1
    return feature_dict 

def bigram_featurize(tokenized_text):
    return {tokenized_text[i]+tokenized_text[i+1]:1 for i in range(len(tokenized_text)-1)}

def unigram(tokenized_text):
    return {tokenized_text[i]:1 for i in range(len(tokenized_text)-1)}
   
bigram_featurize(cleaned_text[0])
```

```python
unigram(cleaned_text[0])
```

```python
train_bound = int(0.7 * len(cleaned_text))
train_bound
```

```python
from sklearn.feature_extraction import DictVectorizer
dv = DictVectorizer(sparse=True)


train_feature_list = [bigram_featurize(x) for x in cleaned_text[:train_bound]]
X_train = dv.fit_transform(train_feature_list)
test_feature_list = [bigram_featurize(x) for x in cleaned_text[train_bound:]]
X_test = dv.transform(test_feature_list)
```

```python
X_train
```

```python
X_test
```

```python
label_list = t_data["label"].apply(make_labels)
```

```python
y_train = label_list[:train_bound]
y_test = label_list[train_bound:]
```

```python
print (len(y_train))
print (len(y_test))
```

```python
collections.Counter(y_train)
```

# 5. Train model ด้วย text และ label จาก Training set



```python
from sklearn.linear_model import LogisticRegression
lr_text_classifier = LogisticRegression()
lr_text_classifier.fit(X_train, y_train)
```

# Evaluation

```python
from sklearn.metrics import accuracy_score
predictions = lr_text_classifier.predict(X_test)
```

```python
print(accuracy_score(y_test, predictions))
```

```python
from sklearn.metrics import confusion_matrix, classification_report
print (classification_report(y_test, predictions))
```

```python
from sklearn.ensemble import RandomForestClassifier

text_classifier = RandomForestClassifier(n_estimators=200, random_state=0)
text_classifier.fit(X_train, y_train)
```

ทดสอบระบบดูว่าแม่นแค่ไหน

```python
from sklearn.metrics import accuracy_score
predictions = text_classifier.predict(X_test)
print(accuracy_score(y_test, predictions))
```

```python
from sklearn.metrics import confusion_matrix, classification_report
print (classification_report(y_test, predictions))
```
