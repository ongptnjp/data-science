# 1. Task Formulation

 Label  

```python
data['labels']
```

```python
valid_labels = set(['', '','', 
                    '', '', ''])
label_row1 = eval(data['labels'][10])
print(label_row1)
```

```python
set(label_row1).intersection(valid_labels)
```

```python
valid_labels = set(['', '','', 
                    '', '', ''])
label_list_list = data['labels'].apply(eval).to_list()
final_label_list = []
for label_list in label_list_list:
    label_set = set(label_list)
    potential_labels = list(label_set.intersection(valid_labels))
    if len(potential_labels) != 1:
        label = ''
    else:
        label = potential_labels[0]
    final_label_list.append(label)
data['label'] = final_label_list
```

```python
mask = data['label'] != ''
mask
```

```python
mask.value_counts()
```

```python
new_data = data[mask]
```

```python
new_data.head()
```

```python
columns_to_keep = ['title tokens', 'label']
new_data = new_data[columns_to_keep]
```

# Train-dev-test split


```python
import numpy as np
num_rows = len(new_data)
first_cutting_point = int(0.8 * num_rows)
second_cutting_point = int(0.9 * num_rows)
shuffled_data = new_data.sample(frac=1)
train, dev, test = np.split(shuffled_data, [first_cutting_point, second_cutting_point])
```

# 2. Feature Extraction

 Feature  bag-of-words


```python
def featurize(token_list):
    features = {}
    for token in token_list:
        features[token] = 1
    features['feature_'] = len(token_list)
    return features
```

```python
print (train['title tokens'][100])
featurize(train['title tokens'][100])
```

#  Sparse Matrix


```python
from sklearn.feature_extraction import DictVectorizer
vectorizer = DictVectorizer(sparse=True)
train_features = train['title tokens'].apply(featurize)
feature_vectors = vectorizer.fit_transform(train_features)
```
